```python
#position choice
```

```python
def position_choice():
    
    choice= "wrong"
    
    while choice not in ['0','1','2']:
        
        choice= input("pick position [0,1,2]:")
        if choice not in ['0','1','2']:
            print("enter valid position.")
    return int(choice)
```

```python
position_choice()
```

```python
#replacement value
```

```python
def replacement_choice(game_list,position):
    
    user_placement= input("type string to place in the position: ")
    
    game_list[position]= user_placement
    
    return game_list
```

```python
replacement_choice(game_list,1)
```

```python
#continuous choice/ game on choice
```

```python
def gameon_choice():
    
    choice= "wrong"
    
    while choice not in ['Y','N']:
        
        choice= input("keep playing? (Y or N): ")
        
        if choice not in ['Y','N']:
            print("invalid choice, choose Y or N")
            
    if choice=="Y":
        return True
    else:
        return False
```

```python
gameon_choice()
```

```python
#compiling all components for choice game
```

```python
game_on=True
game_list=[0,1,2]

while game_on:
    
    display_game(game_list)
    
    position= position_choice()
    
    game_list= replacement_choice(game_list,position)
    
    display_game(game_list)
    
    game_on= gameon_choice()
```
