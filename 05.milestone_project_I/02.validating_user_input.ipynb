```python
#checking data type and range of input
def user_choice():
    
    #initial choice

    choice="wrong"
    acceptable_range=range(0,10)
    within_range=False
    
    
    while choice.isdigit()==False or within_range==False:
        choice= input("input number (0-10):")
        
        if choice.isdigit()==False:
            print("not digit")
            
        if choice.isdigit(): #or if choice.isdigit==True:
            if int(choice) in acceptable_range:
                within_range=True
            else:
                
                print("not in range")
                within_range=False
            
            
    return int(choice)
```
