```python
#sharing same method name to pass diff types of objects
```

```python
class Dog():
    
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        return self.name + " says woof"
```

```python
class Cat():
    
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        return self.name + " says meow"
```

```python
niko=Dog("niko")
felix=Cat("felix")
```

```python
niko.speak()
```

```python
print(felix.speak())
```

```python
#demonstrating polymorphism
```

```python
for pet in [niko,felix]:
    print(type(pet))
    print(pet.speak())
```

```python
def pet_speaks(pet):
    print(pet.speak())
```

```python
pet_speaks(niko)
```

```python
pet_speaks(felix)
```

```python
#i.e we obtained object specific  responses
```

###more common practice: using abstract classes and inheritance

```python
#abstract class: no instances, serves as base class
```

```python
class Animal():
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        raise NotImplementedError("subclass must implement this abstract method")
        #it is expected that base class inherits Animal class and overwrites th espeak method
```

```python
my_animal=Animal("bobby")
```

```python
my_animal.speak()
```

```python
#so we create derived class

class Dog(Animal):
    
    def speak(self):
        return self.name+" says  bhau bhau"
```

```python
class Cat(Animal):
    
    def speak(self):
        return self.name+" says  meowww"
```

```python
gorey=Dog("gorey")
```

```python
biralo=Cat("biralo")
```

```python
gorey.speak()
```

```python
biralo.speak()
```
