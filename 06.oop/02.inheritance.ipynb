```python
#base class

class Animal():
    
    def __init__(self):
        print("animal created")
        
    def identity(self):
        print("i am an animal")
        
    def eat(self):
        print("i eat")
```

```python
my_animal=Animal()
```

```python
my_animal.eat()
```

```python
my_animal.identity()
```

```python
#derived class

class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("dog created")
        
    
```

```python
my_dog=Dog()
```

```python
my_dog.eat()
```

```python
my_dog.identity()
```

```python
#rewriting derived class method
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("dog created")
    def identity(self):
        print("i am a dog")   
 

my_dog=Dog()        
my_dog.identity()
    
```

```python
#aadding methods in base class
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("dog created")
    def identity(self):
        print("i am a dog") 
    def bark(self):
        print("woof!")
        
my_dog=Dog()        
my_dog.bark()
 
```
