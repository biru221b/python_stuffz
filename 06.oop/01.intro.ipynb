```python
class Dog():
  
    species="mammal"
    
    def __init__(self, breed,name):
        self.name=name
        self.breed=breed
        
    #operations/actions->methods
    def bark(self):
        print("woof!, ma kukur ho!!!")
        
my_dog= Dog(name="kaaley", breed="bhusyaha")
```

```python
my_dog.name
```

```python
my_dog.bark()
```

```python
class Dog():
  
    species="mammal"
    
    def __init__(self, breed,name):
        self.name=name
        self.breed=breed
        
    #operations/actions->methods
    def bark(self, number):
        print("woof!, ma {} kukur ho!!! mero {} chhaura chhan".format(self.name,number))
        
        #dont use .format(my_dog.name, number)),
        #Problems:
        #my_dog is a global variable, not guaranteed to exist when the method runs.
        #Methods should always use self to refer to the object that called them.
        #If you create another dog object, the method would still incorrectly refer to my_dog

        
my_dog= Dog(name="kaaley", breed="bhusyaha")
```

```python
my_dog.bark(10)
```

```python
class Circle():
    
    #class object attr
    pi= 3.1415
    
    def __init__(self,rad=1):
        self.rad=rad
        #attr doesnt need to be defined 
        self.area= (rad**2) * self.pi  #or Circle.pi
        
    def get_circum(self):
        return self.rad* self.pi *2
    
    
```

```python
my_circle=Circle()
```

```python
my_circle.pi
```

```python
my_circle.rad
```

```python
my_circle.get_circum()
```

```python
my_circle.area
```
