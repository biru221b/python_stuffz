## range,map,filter quick recap

map(function, iterable)

```python
def square(x):
    return x * x

numbers = [1,2,3,4]

result = map(square, numbers)

print(list(result))
```

filter(function, iterable)

```python
def is_even(x):
    return x % 2 == 0

numbers = [1,2,3,4,5,6]

result = filter(is_even, numbers)

print(list(result))
```

 Generator functions allow us to write a function that can send back a value and then later resume to pick up where it left off. This type of function is a generator in Python, allowing us to generate a sequence of values over time.

In most aspects, a generator function will appear very similar to a normal function. The main difference is when a generator function is compiled they become an object that supports an iteration protocol. That means when they are called in your code they don't actually return a value and then exit. Instead, generator functions will automatically suspend and resume their execution and state around the last point of value generation. The main advantage here is that instead of having to compute an entire series of values up front, the generator computes one value and then suspends its activity awaiting the next instruction. This feature is known as state suspension.

```python
#generating  cubes using normal function
```

```python
def create_cubes(n):
    result=[]
    for x in range (n):
        result.append(x**3)
        
    return result
```

```python
create_cubes(5)
```

```python
#using generator
```

```python
def create_cubes(n):
    
    for x in range (n):
        yield x**3
        
```

```python
create_cubes(5)
```

```python
for x in create_cubes(5):
    print (x)
```

```python
list(create_cubes(5))
```

```python
#fibonacci sequence generator
```

```python
def gen_fibon(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a,b= b,a+b
```

```python
list(gen_fibon(10))
```

# next() and iter() built-in functions

```python
#The next() function allows us to access the next element in a sequence. 
```

```python

def simple_gen():
    for x in range(3):
        yield x
```

```python
list(simple_gen())
```

```python

# Assign simple_gen 
g = simple_gen()
```
