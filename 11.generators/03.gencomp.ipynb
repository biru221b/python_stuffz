## generator comprehension

generator comprehension (or generator expression)  <br>

It creates a generator object that produces values lazily (one at a time) instead of creating a full list in memory.

```python
#list comprehension
```

[item for item in my_list if item > 3] 

```python
#generator comprehension
```

(item for item in my_list if item > 3)

```python
#square numbers
```

```python
nums = [1,2,3,4]

gen = (x*x for x in nums)

for i in gen:
    print(i)
```

```python
#multiple conditions
```

```python
nums = range(20)

gen = (x for x in nums if x % 2 == 0 if x > 10)

print(list(gen))
```

```python
#nested loop
```

```python
pairs = ((x, y) for x in range(3) for y in range(3))

print(list(pairs))
```

```python
#generator inside function
```

```python
nums = [1,2,3,4,5]

total = sum(x*x for x in nums)

print(total)
```

```python
#conditional transformation
```

```python
nums = range(10)

gen = (x*x if x%2==0 else x for x in nums)

print(list(gen))
```
