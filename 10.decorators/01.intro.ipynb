# Functions within functions

```python
def hello(name="jose"):
    print("the hello() func is executed")
    
    def greeting():
        return "this is greeting() func inside hello"
    
    def wel():
        print("welcome")
    
    print(greeting())  #to print greet, instead greet() wont print as it is not called
    print(wel())   #returns none as well if nothing is returned
```

```python
hello()
```

```python
wel()
```

```python
def hello(name="jose"):
    print("the hello() func is executed")
    
    def greeting():
        return "\t this is greeting() func inside hello"
    
    def wel():
        print("welcome")
    
    print("returning a function")
    
    if name=="jose":
        return greeting
    else:
        return wel
        
```

```python
my_new_func=hello("jose")
```

```python
my_new_func

#points to the greeting func inside hello func as name is jose
```

```python
my_new_func()
```

```python
print(my_new_func())
```

# #decorator

```python
def new_decorator(original_func):
    
    def wrap_func():
        
        print("some code before oroginal function")
        
        original_func()
        
        print("some code after original func")
        
    return wrap_func
```

```python
def func_needs_decorator():
    print("i want to be decorated")
```

```python
func_needs_decorator()
```

```python
decorated_func=new_decorator(func_needs_decorator)
```

```python
decorated_func()
```

```python
# @new_decorator
def func_needs_decorator():
    print("i want to be decorated")
```

```python
@new_decorator
def func_needs_decorator():
    print("i want to be decorated")
    
```

```python
func_needs_decorator()
```
