```python
try:
    #want to attempt this code
    result= 10+10
    
except:
    print("looks like  you aren't adding properly")
else:
    print("all went well")
    print(result)
```

```python
try:
    #want to attempt this code
    result= 10+'10'
    
except:
    print("looks like  you aren't adding properly")
else:
    print("all went well")
    print(result)
```

```python
#errors reference from https://docs.python.org/3/library/exceptions.html#bltin-exceptions
```

```python
try:
    f= open("randomfile", "w")
    f.write("write a test line")
except TypeError:
    print("type error")
except OSError:
    print("there is an os error")

finally:
    print("i always run")
```

```python
try:
    f= open("randomfile", "r")
    f.write("write a test line")
except TypeError:
    print("type error")
except OSError:
    print("there is an os error")

finally:
    print("i always run")
```

```python
#sorting out excptions:
try:
    f= open("randomfile", "r")
    f.write("write a test line")
except TypeError:
    print("type error")
except:
    print("all other excptions")

finally:
    print("i always run")
```

```python
def ask_intro():
    try:
        result=int(input("please provide number: "))
    except:
        print("not a number mf")
    finally:
        print("thanks")
```

```python
ask_intro()
```

```python
ask_intro()
```

```python
#looping until no exception
```

```python
def ask_intro():
    
    while True:
        try:
            result=int(input("please provide number: "))
        except:
            print("not a number mf")
            continue
            
        else:
            print("good ")
            break
        finally:
            print("thanks")
```

```python
ask_intro()
```
